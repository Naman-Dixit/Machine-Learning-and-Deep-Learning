```python
import builtins

# Input list from the user
elements = list(map(int, builtins.input("Enter numbers separated by space: ").split()))

# Function to check even or odd
def even_or_odd(arr):
    for num in arr:
        if num % 2 == 0:
            print(num, "is Even")
        else:
            print(num, "is Odd")

# Function call
even_or_odd(elements)
```

```python
import builtins

# Input list from the user
elements = list(map(int, builtins.input("Enter numbers separated by space: ").split()))

def MinMaxSumAvg(arr):
    total = 0                     # Do NOT use name 'sum'
    minimum = arr[0]              # Initialize min
    maximum = arr[0]              # Initialize max

    for num in arr:
        total += num
        if num < minimum:
            minimum = num
        if num > maximum:
            maximum = num

    print("Min:", minimum)
    print("Max:", maximum)
    print("Sum:", total)
    print("Avg:", total / len(arr))

MinMaxSumAvg(elements)
```
